Importando as informações sobre o acervo de livros

```python
informacoes_acervo = pd.read_parquet("https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/blob/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet?raw=true", engine="pyarrow")
```

Concatenando as bases de dados em uma única tabela

```python
bases_emprestimos = [
    base_2010_1,
    base_2010_2,
    base_2011_1,
    base_2011_2,
    base_2012_1,
    base_2012_2,
    base_2013_1,
    base_2013_2,
    base_2014_1,
    base_2014_2,
    base_2015_1,
    base_2015_2,
    base_2016_1,
    base_2016_2,
    base_2017_1,
    base_2017_2,
    base_2018_1,
    base_2018_2,
    base_2019_1,
    base_2019_2,
    base_2020_1,
]

base_emprestimo = pd.concat(bases_emprestimos)
```

Mesclando as duas tabelas

```python
informacoes_emprestimos = pd.merge(base_emprestimo, informacoes_acervo, on='codigo_barras', how='left')
```

Removendo as duplicatas da base de dados

```python
informacoes_emprestimos.drop_duplicates(inplace=True)
```

Preenchendo os dados nulos da coluna "Matricula_ou_siape"

```python
print(informacoes_emprestimos.isnull().sum())
```

```python
informacoes_emprestimos["matricula_ou_siape"] = informacoes_emprestimos["matricula_ou_siape"].fillna("Não identificado")
```

Usando 'bfill' e 'ffill' para prencher os dados faltantes nas colunas de tempo

```python
informacoes_emprestimos["data_devolucao"] = informacoes_emprestimos["data_devolucao"].ffill()
informacoes_emprestimos["data_renovacao"] = informacoes_emprestimos["data_renovacao"].ffill()
informacoes_emprestimos["data_renovacao"] = informacoes_emprestimos["data_renovacao"].bfill()
```

Descartando as linhas vazias referente ao acervo

```python
df_nulo = informacoes_emprestimos[informacoes_emprestimos.isna().any(axis=1)]
print(df_nulo)
```

```python
informacoes_emprestimos = informacoes_emprestimos.dropna(axis=0)
```

```python
print(informacoes_emprestimos.isnull().sum())
```

Excluindo a coluna 'registro_sistema'

```python
del informacoes_emprestimos["registro_sistema"]
```

Criando uma nova coluna para a base de dados

```python
import numpy as np

condicoes = [
    ((informacoes_emprestimos["localizacao"] >= 0) & (informacoes_emprestimos["localizacao"] <= 99)),
    ((informacoes_emprestimos["localizacao"] >= 100) & (informacoes_emprestimos["localizacao"] <= 199)),
    ((informacoes_emprestimos["localizacao"] >= 200) & (informacoes_emprestimos["localizacao"] <= 299)),
    ((informacoes_emprestimos["localizacao"] >= 300) & (informacoes_emprestimos["localizacao"] <= 399)),
    ((informacoes_emprestimos["localizacao"] >= 400) & (informacoes_emprestimos["localizacao"] <= 499)),
    ((informacoes_emprestimos["localizacao"] >= 500) & (informacoes_emprestimos["localizacao"] <= 599)),
    ((informacoes_emprestimos["localizacao"] >= 600) & (informacoes_emprestimos["localizacao"] <= 699)),
    ((informacoes_emprestimos["localizacao"] >= 700) & (informacoes_emprestimos["localizacao"] <= 799)),
    ((informacoes_emprestimos["localizacao"] >= 800) & (informacoes_emprestimos["localizacao"] <= 899)),
    ((informacoes_emprestimos["localizacao"] >= 900) & (informacoes_emprestimos["localizacao"] <= 999))
]

opcoes = [
    "Generalidades. Ciência e conhecimento.",
    "Filosofia e psicologia.",
    "Religião.",
    "Ciências sociais.",
    "Classe vaga. Provisioramente não ocupada.",
    "Matemática e ciências naturais.",
    "Ciências aplicadas.",
    "Belas artes.",
    "Linguagem. Língua. Linguística.",
    "Geografia. Biografia. História."
]

informacoes_emprestimos["genero"] = np.select(condicoes, opcoes, default="Outros")
```

Modificando a coluna de data da base de dados

```python
informacoes_emprestimos["data_emprestimo"] = pd.to_datetime(informacoes_emprestimos["data_emprestimo"])
```

Criando a variável para armazenar os dados para os gráficos dos anos

```python
emprestimo_por_ano = informacoes_emprestimos["data_emprestimo"].dt.year.value_counts().reset_index()
```

```python
emprestimo_por_ano.columns = ['Ano', 'Quantidade']
emprestimo_por_ano = emprestimo_por_ano.sort_values(by="Ano")
```

Criando o gráfico de empréstimo por ano

```python
emprestimo_por_ano.plot(kind="line", x="Ano", y="Quantidade", figsize=(12, 6));
```
