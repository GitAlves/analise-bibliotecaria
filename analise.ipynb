```python
dfs = [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12, df_13, df_14, df_15, df_16, df_17, df_18, df_19, df_20, df_21]
dfs_concatenados = pd.concat(dfs)
print(dfs_concatenados)
```

```python
dfs_concatenados.value_counts()
```

```python
df_concatenados = dfs_concatenados.drop_duplicates()

dfs_concatenados.value_counts()
```

```python
esquerdo = dfs_concatenados
direito = df_acervo
df_comum = pd.merge(esquerdo, direito, on="codigo_barras")
```

```python
print(df_comum)
```

```python
print(df_comum.isnull().sum())
```

```python
moda = df_comum['matricula_ou_siape'].mode()[0]

df_comum['matricula_ou_siape'] = df_comum['matricula_ou_siape'].fillna(moda)
```

```python
print(df_comum.isnull().sum())
```
